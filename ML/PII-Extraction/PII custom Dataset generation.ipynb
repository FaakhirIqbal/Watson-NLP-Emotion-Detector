### 1. Create PII Driving License Number  Dataset from Faker library

Let's generate PII data for USA state Driving License number using Faker.

```python
#Prepared funtion for USA state DLN
from faker import Faker
import random

fake = Faker()

#Colorado
def generate_driving_license_Colarado():
    license_number = fake.numerify('##-###-####')
    name = fake.name()
    state = "Colarado"
    return license_number, name, state

#Alaska & Alabama
def generate_driving_license_Alaska():
    license_number = fake.numerify('#######')
    name = fake.name()
    state = random.choice(["Alaska","Alabama"])
    return license_number, name, state

#Arkansas & South Carolina
def generate_driving_license_SCarolina():
    license_number = f"{9}{fake.numerify('########')}"
    name = fake.name()
    state = random.choice(["Arkansas","South Carolina"])
    return license_number, name, state

#California
def generate_driving_license_California():
    license_number = f"{'A'}{fake.numerify('########')}"
    name = fake.name()
    state = random.choice(["California"])
    return license_number, name, state

#Hawaii
def generate_driving_license_Hawaii():
    license_number = f"{'H'}{fake.numerify('########')}"
    name = fake.name()
    state = random.choice(["Hawaii"])
    return license_number, name, state

#New York
def generate_driving_license_New_York():
    license_number = fake.numerify('### ### ###')
    name = fake.name()
    state = random.choice(["New York"])
    return license_number, name, state

#North Carolina
def generate_driving_license_NCarolina():
    license_number = fake.numerify('############')
    name = fake.name()
    state = random.choice(["North Carolina"])
    return license_number, name, state

#California
def generate_driving_license_California():
    license_number = f"{'A'}{fake.numerify('########')}"
    name = fake.name()
    state = random.choice(["California"])
    return license_number, name, state

#Texas
def generate_driving_license_Texas():
    license_number = f"{'A'}{fake.numerify('########')}"
    name = fake.name()
    state = random.choice(["California"])
    return license_number, name, state
```

```python
def format_data():
    State_DLN = random.choice([generate_driving_license_Colarado(), generate_driving_license_Alaska(), generate_driving_license_SCarolina(), generate_driving_license_California(), generate_driving_license_Hawaii(), generate_driving_license_New_York(), generate_driving_license_NCarolina(), generate_driving_license_Texas()])
    driving_license, name, state = State_DLN

    text_1 = "My name is %s I belong to the %s , My Driving License number is %s." %(name, state, driving_license)
    text_2 = "I am %s. %s this is my driving license number. I am from %s state." %(name, driving_license, state)
    text_3 = "Hello, My self %s, I am living in %s and my driving License number is %s"  %(name, state, driving_license)
    text = random.choice([text_1, text_2, text_3])
    
    
    name_begin = text.find(name)
    name_end = name_begin + len(name)
    state_begin = text.find(state)
    state_end = state_begin + len(state)
    driving_license_begin = text.find(driving_license)
    driving_license_end = driving_license_begin + len(driving_license)
    
    
    data = {
                "text": text,
                "mentions": [
                    {
                        "location": {
                            "begin": name_begin,
                            "end": name_end
                        },
                        "text": name,
                        "type": "Name"
                    },
                    {
                        "location": {
                            "begin": state_begin,
                            "end": state_end
                        },
                        "text": state,
                        "type": "state"
                    },
                    {
                        "location": {
                            "begin": driving_license_begin,
                            "end": driving_license_end
                        },
                        "text": driving_license,
                        "type": "driving_license_number"
                    },
                ]   
            }
    
    return data
```

### 2. Create PII Dataset from Faker library

Let's generate PII data from a mix of Faker and custom generators.

```python
faker = Faker(locale='en_US')
name_faker = Faker(locale=['en_TH', 'en_IE', 'en_NZ', 'en_IN', 'en_US'])
```

```python
def generate_race():
    races = ['American Indian', 'Alaska Native', 'Asian', 'African American', 'Black', 'Hispanic', 'Latino', 'Native Hawaiian', 'Pacific Islander', 'White']
    return random.choice(races)
```

```python
def generate_credit_card_company():
    credit_card_companies = ['American Express', 'Discover', 'MasterCard', 'Visa']
    return random.choice(credit_card_companies)
```

```python
def format_data():
    name_non = name_faker.prefix_nonbinary() + " " + name_faker.first_name_nonbinary() + " " + name_faker.last_name_nonbinary() + " " + name_faker.suffix_nonbinary()
    name_m = name_faker.prefix_male() + " " + name_faker.first_name_male() + " " + name_faker.last_name_male() + " " + name_faker.suffix_male()
    name_f = name_faker.prefix_female() + " " + name_faker.first_name_female() + " " + name_faker.last_name_female() + " " + name_faker.suffix_male()
    race = generate_race()
    phone = faker.phone_number()
    email = faker.ascii_free_email()
    address = faker.address()
    ssn = faker.ssn()
    ccc = generate_credit_card_company()
    ccn = faker.credit_card_number()
    salary = faker.pricetag()
    
    text_non = "My name is %s, I'm a gender neutral person. I belong to the %s race. You can call me at %s, or email me at %s, or visit my home at %s. My social security number is %s. Here's the number to my %s credit card, %s. I make a salary of %s per year." %(name_non, race, phone, email, address, ssn, ccc, ccn, salary)
    text_m = "My name is %s, I'm a male. I belong to the %s race. You can call me at %s, or email me at %s, or visit my home at %s. My social security number is %s. Here's the number to my %s credit card, %s. I make a salary of %s per year." %(name_m, race, phone, email, address, ssn, ccc, ccn, salary)
    text_f = "My name is %s, I'm a female. I belong to the %s race. You can call me at %s, or email me at %s, or visit my home at %s. My social security number is %s. Here's the number to my %s credit card, %s. I make a salary of %s per year." %(name_f, race, phone, email, address, ssn, ccc, ccn, salary)
    
    text = random.choice([text_non, text_m, text_f])
    
    if text == text_non:
        name = name_non
        gender = "gender neutral person"
    elif text == text_m:
        name = name_m
        gender = "male"
    elif text == text_f:
        name = name_f
        gender = "female"
    
    name_begin = text.find(name)
    name_end = name_begin + len(name)
    race_begin = text.find(race)
    race_end = race_begin + len(race)
    gender_begin = text.find(gender)
    gender_end = gender_begin + len(gender)
    phone_begin = text.find(phone)
    phone_end = phone_begin + len(phone)
    email_begin = text.find(email)
    email_end = email_begin + len(email)
    address_begin = text.find(address)
    address_end = address_begin + len(address)
    ssn_begin = text.find(ssn)
    ssn_end = ssn_begin + len(ssn)
    ccn_begin = text.find(ccn)
    ccn_end = ccn_begin + len(ccn)
    salary_begin = text.find(salary)
    salary_end = salary_begin + len(salary)
    
    data = {
                "text": text,
                "mentions": [
                    {
                        "location": {
                            "begin": name_begin,
                            "end": name_end
                        },
                        "text": name,
                        "type": "Name"
                    },
                    {
                        "location": {
                            "begin": race_begin,
                            "end": race_end
                        },
                        "text": name,
                        "type": "Race"
                    },
                    {
                        "location": {
                            "begin": gender_begin,
                            "end": gender_end
                        },
                        "text": gender,
                        "type": "Gender"
                    },
                    {
                        "location": {
                            "begin": phone_begin,
                            "end": phone_end
                        },
                        "text": phone,
                        "type": "PhoneNumber"
                    },
                    {
                        "location": {
                            "begin": email_begin,
                            "end": email_end
                        },
                        "text": email,
                        "type": "Email"
                    },
                    {
                        "location": {
                            "begin": address_begin,
                            "end": address_end
                        },
                        "text": address,
                        "type": "Address"
                    },
                    {
                        "location": {
                            "begin": ssn_begin,
                            "end": ssn_end
                        },
                        "text": ssn,
                        "type": "SocialSecurityNumber"
                    },
                    {
                        "location": {
                            "begin": ccn_begin,
                            "end": ccn_end
                        },
                        "text": ccn,
                        "type": "CreditCardNumber"
                    },
                    {
                        "location": {
                            "begin": salary_begin,
                            "end": salary_end
                        },
                        "text": salary,
                        "type": "Salary"
                    }
                ]   
            }
    
    return data
```

```python
def format_data(df, name_col, ssn_col, ccn_col):  
    import random
    
    train_list = []
    for i in range(1, len(df)):
        name = df[name_col][i] 
        ssn = str(df[ssn_col][i])
        ccn = str(df[ccn_col][i])
        
        text1 = "My name is %s, and my social security number is %s. Here's the number to my Visa credit card, %s" % (name, ssn, ccn)
        text2 = "%s is my social security number. The name on my American Express card %s is %s." % (ssn, ccn, name)

        text = random.choice([text1, text2])

        name_begin = text.find(name)
        name_end = text.find(name) + len(name)
        ssn_begin = text.find(ssn)
        ssn_end = text.find(ssn) + len(ssn)
        ccn_begin = text.find(ccn)
        ccn_end = text.find(ccn) + len(ccn)

        data = {
                    "text": text,
                    "mentions": [
                        {
                            "location": {
                                "begin": name_begin,
                                "end": name_end
                            },
                            "text": name,
                            "type": "Name"
                        },
                        {
                            "location": {
                                "begin": ssn_begin,
                                "end": ssn_end
                            },
                            "text": ssn,
                            "type": "SocialSecurityNumber"
                        },
                        {
                            "location": {
                                "begin": ccn_begin,
                                "end": ccn_end
                            },
                            "text": ccn,
                            "type": "CreditCardNumber"
                        }
                    ]   
                }

        train_list.append(data)
    return train_list
```
